```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_excel("data.xlsx")
df = df.set_index(df.columns[0])
```

```python
import pandas as pd
import numpy as np

# Son iki s√ºtunun adlarƒ±nƒ± al
son_sutunlar = df.columns[-2:]

# Her bir s√ºtun i√ßin i≈ülem yap
for sutun in son_sutunlar:
    # Bo≈ü h√ºcreleri NaN olarak i≈üaretle
    df[sutun] = df[sutun].replace('', np.nan)
    
    # Bo≈ü h√ºcreleri doldur
    df[sutun] = df[sutun].fillna(method='ffill', limit=10)  # √ñnceki 10 satƒ±rƒ±n ortalamasƒ±yla doldur
```

```python
df.plot(kind='box',figsize=(15,10),subplots=True,layout=(3,4))
plt.show()
```

```python
plt.figure(figsize=(20,20)) 
p=sns.heatmap(df.corr(), annot=True,cmap='RdYlGn',square=True) 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Veri seti ile pairplot olu≈ütur
g = sns.pairplot(df, kind="scatter")

for ax in g.axes.flatten():
    # X eksen etiketlerini 90 derece d√∂nd√ºr ve b√ºy√ºt
    ax.set_xlabel(ax.get_xlabel(), rotation=90, fontsize=20)
    # Y eksen etiketlerini 0 derece d√∂nd√ºr ve b√ºy√ºt
    ax.set_ylabel(ax.get_ylabel(), rotation=0, fontsize=20)
    # Y eksen etiketlerinin hizalanmasƒ±nƒ± saƒüa ayarla
    ax.yaxis.get_label().set_horizontalalignment('right')

# Subplotlar arasƒ±ndaki bo≈üluƒüu artƒ±r ve ba≈ülƒ±ƒüƒ± ayarla
plt.tight_layout()
plt.show()
```
