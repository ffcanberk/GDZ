```python
"""
Dataset Description

Days — Kesintinin gerçekleştiği tarih
ilce — Kesintinin gerçekleştiği il-ilçe. Örnek: “manisa-akhisar”
Sum of Planlı - günlük Planlı kesinti sayısı
Sum of Plansız — günlük Plansız kesinti sayısı (Tahmin edilecek hedef değişken 🎯)

 avg. t_2m:C — günlük Sıcaklık, Celcius
 avg. effective_cloud_cover:p — günlük Bulutluluk
 avg. global_rad:W — günlük Güneşlilik değeri, Watt
 avg. relative_humidity_2m:p — günlük Bağıl nem
 avg. wind_dir_10m:d — günlük Rüzgar yönü
 avg. wind_speed_10m:ms — günlük Rüzgar hızı, m/s
 avg. prob_precip_1h:p — günlük Yağış
 avg. t_apparent:C — günlük Hissedilen sıcaklık

"""
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_excel("data.xlsx")
df = df.set_index(df.columns[0])
```

```python
import pandas as pd
import numpy as np

# Son iki sütunun adlarını al
son_sutunlar = df.columns[-2:]

# Her bir sütun için işlem yap
for sutun in son_sutunlar:
    # Boş hücreleri NaN olarak işaretle
    df[sutun] = df[sutun].replace('', np.nan)
    
    # Boş hücreleri doldur
    df[sutun] = df[sutun].fillna(method='ffill', limit=10)  # Önceki 10 satırın ortalamasıyla doldur
```

```python
df.plot(kind='box',figsize=(15,10),subplots=True,layout=(3,4))
plt.show()
```

```python
plt.figure(figsize=(20,20)) 
p=sns.heatmap(df.corr(), annot=True,cmap='RdYlGn',square=True) 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Veri seti ile pairplot oluştur
g = sns.pairplot(df, kind="scatter")

for ax in g.axes.flatten():
    # X eksen etiketlerini 90 derece döndür ve büyüt
    ax.set_xlabel(ax.get_xlabel(), rotation=90, fontsize=20)
    # Y eksen etiketlerini 0 derece döndür ve büyüt
    ax.set_ylabel(ax.get_ylabel(), rotation=0, fontsize=20)
    # Y eksen etiketlerinin hizalanmasını sağa ayarla
    ax.yaxis.get_label().set_horizontalalignment('right')

# Subplotlar arasındaki boşluğu artır ve başlığı ayarla
plt.tight_layout()
plt.show()
```
